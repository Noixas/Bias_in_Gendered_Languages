```python
def load_vec(emb_path, nmax=50000):
    vectors = []
    word2id = {}
    with io.open(emb_path, 'r', encoding='utf-8', newline='\n', errors='ignore') as f:
        next(f)
        for i, line in enumerate(f):
            word, vect = line.rstrip().split(' ', 1)
            vect = np.fromstring(vect, sep=' ')
            assert word not in word2id, 'word found twice'
            vectors.append(vect)
            word2id[word] = len(word2id)
            if len(word2id) == nmax:
                break
    id2word = {v: k for k, v in word2id.items()}
    embeddings = np.vstack(vectors)
    return embeddings, id2word, word2id
```

```python
def save_vec(embeddings, id2word, emb_path):
    with io.open(emb_path, 'w', encoding='utf-8', newline='\n', errors='ignore') as f:
        f.write('50000 300\n')
        for i in range(len(embeddings)):
            f.write(id2word[i]+' ')
            for number in embeddings[i]:
                f.write(str(number)+' ')
            f.write('\n')
            
#ori_offset_debiased_ES_embeddings_path= './data/ori_offset_debiased_ES_embeddings.txt'   
#save_vec(ori_offset_debiased_ES_embeddings, spanish_id2word, emb_path=ori_offset_debiased_ES_embeddings_path)
```

# Get nearest neighbors

```python
def get_nn(word, src_emb, src_id2word, tgt_emb, tgt_id2word, K=5):
    print("Nearest neighbors of \"%s\":" % word)
    word2id = {v: k for k, v in src_id2word.items()}
    word_emb = src_emb[word2id[word]]
    scores = (tgt_emb / np.linalg.norm(tgt_emb, 2, 1)[:, None]).dot(word_emb / np.linalg.norm(word_emb))
    k_best = scores.argsort()[-K:][::-1]
    #for i, idx in enumerate(k_best):
        #print('%.4f - %s' % (scores[idx], tgt_id2word[idx]))
    nn = []
    for idx in k_best:
        nn.append(tgt_id2word[idx])
    return(nn)
```

# Calculate cosine similarity

```python
def cos_sim(emb1, emb2):
    return((emb1/np.linalg.norm(emb1)).dot(emb2/np.linalg.norm(emb2)))

def get_cos_sim(word1, src_emb, src_id2word, word2, tgt_emb, tgt_id2word):
    word2id_1 = {v: k for k, v in src_id2word.items()}
    emb_1 = src_emb[word2id_1[word1]]
    word2id_2 = {v: k for k, v in tgt_id2word.items()}
    emb_2 = tgt_emb[word2id_2[word2]] 
    return((emb_1/np.linalg.norm(emb_1)).dot(emb_2/np.linalg.norm(emb_2)))

def get_cos_sim_neutral(word, emb, id2word):
    word2id = {v: k for k, v in id2word.items()}
    word_emb = emb[word2id[word]]
    return((neutral_embedding/np.linalg.norm(neutral_embedding)).dot(word_emb/np.linalg.norm(word_emb)))
```

# Get Gender Directions

```python
from sklearn.decomposition import PCA

def doPCA(pairs, embedding, word2id, num_components = 10):
    matrix = []
    for a, b in pairs:
        center = (embedding[word2id[a]] + embedding[word2id[b]]/2)
        matrix.append(embedding[word2id[a]] - center)
        matrix.append(embedding[word2id[b]] - center)
    matrix = np.array(matrix)
    pca = PCA(n_components = num_components)
    pca.fit(matrix)
    # bar(range(num_components), pca.explained_variance_ratio_)
    return pca

def doPCA_multi(pairs, embedding1, embedding2, word2id1, word2id2, num_components = 10):
    matrix = []
    count = 0
    for a, b in pairs:
        if count < 10:
            embedding = embedding1
            word2id = word2id1
        else:
            embedding = embedding2
            word2id = word2id2
        center = (embedding[word2id[a]] + embedding[word2id[b]]/2)
        matrix.append(embedding[word2id[a]] - center)
        matrix.append(embedding[word2id[b]] - center)
        count += 1
    matrix = np.array(matrix)
    pca = PCA(n_components = num_components)
    pca.fit(matrix)
    # bar(range(num_components), pca.explained_variance_ratio_)
    return pca
```

```python
gender_direction = ES_semantic_gender_direction
word_embeddings = spanish_embeddings
emb_word2id = spanish_word2id

def get_most_biased_words(embeddings, word2id, direction, topK=500):
    dot_products = {}
    for i in range(len(embeddings)):
        dot_products[i] = (embeddings[i]/np.linalg.norm(embeddings[i])).dot(direction/np.linalg.norm(direction))
    sorted_dot_products = sorted(dot_products.items(), key=lambda kv: kv[1])
    smallest = sorted_dot_products[0:topK]
    largest = sorted_dot_products[50000:(50000-topK-1):-1]
    return smallest, largest
    
smallest_500, largest_500 = get_most_biased_words(word_embeddings, emb_word2id, gender_direction)
  
gender_direction = ES_grammar_gender_direction_3000
smallest_500_grammar, largest_500_grammar = get_most_biased_words(word_embeddings, emb_word2id, gender_direction)

```

```python
import math
from itertools import combinations, filterfalse
import logging
logger = logging.getLogger()
logger.setLevel(logging.DEBUG)


random.seed(520)
def _weat_pval(embeddings, word2id, tw1, tw2, aw1, aw2, modified=1):
    test_stat = modified_WEAT(embeddings, word2id, list(zip(tw1,tw2)), aw1, aw2)[modified]
    print(test_stat)
    observed_over = []
    all_targets = tw1 + tw2
#     logging.info('Calculating p val over {2:.0f} length-{0} partitions of {1} words'.format(
#         len(tw1), len(all_targets), 
#         math.factorial(len(all_targets))/
#         (math.factorial(len(tw1))*(math.factorial(len(all_targets)-len(tw1))))))
    if modified == 0:
        for c in combinations(all_targets, len(tw1)):
            not_c = filterfalse(lambda x: x in c, all_targets)
            stat = modified_WEAT(embeddings, word2id, list(zip(c, not_c)), aw1, aw2)[modified]
            observed_over.append(stat > test_stat)
            if len(observed_over) % 1000 == 0:
                logging.info("  Iteration {0}".format(len(observed_over)))
    elif modified == 1:
        for i in range(1000):
            c = random.sample(all_targets, len(tw1))
            not_c = filterfalse(lambda x: x in c, all_targets)
            stat = modified_WEAT(embeddings, word2id, list(zip(c, not_c)), aw1, aw2)[modified]
            observed_over.append(stat > test_stat)
            if len(observed_over) % 1000 == 0:
                logging.info("  Iteration {0}".format(len(observed_over)))
    return(np.mean(observed_over))

```

```python
def doPCA_(pairs, num_components = 13):
    matrix = []
    counter = 0
    for a, b in pairs:
        if counter <= 9:
            center = (original_embeddings[original_word2id[a]] + original_embeddings[original_word2id[b]])/2
            matrix.append(original_embeddings[original_word2id[a]] - center)
            matrix.append(original_embeddings[original_word2id[b]] - center)
        else:
            #break
            center = (spanish_embeddings[spanish_word2id[a]] + spanish_embeddings[spanish_word2id[b]])/2
            matrix.append(spanish_embeddings[spanish_word2id[a]] - center)
            matrix.append(spanish_embeddings[spanish_word2id[b]] - center)
        counter+= 1
    matrix = np.array(matrix)
    pca = PCA(n_components = num_components)
    pca.fit(matrix)
    # bar(range(num_components), pca.explained_variance_ratio_)
    return pca.explained_variance_ratio_
doPCA(defs)

plt.bar(range(1,14),doPCA(defs))
plt.show()
```

```python
from sklearn.manifold import TSNE
import numpy as np
import matplotlib.pyplot as plt
import math
import nltk

adjs = ['JJ', 'JJR', 'JJS']
nns = ['NN', 'NNS', 'NNP', 'NNPs']

def fit_tsne(values):
    if not values:
        return
    mat = np.array(values)
    model = TSNE(n_components=2, random_state=0, learning_rate=150, init='pca')
    fitted = model.fit_transform(mat)
    return fitted

def plot_word(word1, words, fitted, cmap, sims):
    plt.scatter(fitted[:,0], fitted[:,1], alpha=0)
    # plt.suptitle("[topk for %s]" % word1, fontsize=20, y=0.1, color='black')
    plt.axis('off')
    annotations = []
    isArray = type(word1) == list
    for i in range(len(words)):
        pt = fitted[i]
        ww= words[i]
        #print(i)
        color = cmap[i]
        sizing = sims[words[i]] * 60
        plt.text(pt[0], pt[1], ww, color=color, size=int(sizing))
    return annotations

def plot_one_word(word, embedding, wordset = [], topk=10, flag = '', w2v = False):
    plt.close('all')
    values = []
    tops = 300
    if w2v == True:
        simr = embedding.most_similar(positive=[word], topn = tops)
    else:
        simr = embedding.closest(word, tops)
    
    if flag == 'adj':
        simr = [x for x in simr if nltk.pos_tag([x[1].strip()])[0][1] in adjs and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
    elif flag == 'nn':
        simr = [x for x in simr if nltk.pos_tag([x[1].strip()])[0][1] in nns and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
    else:
        if w2v == True:
            simr = [x for x in simr if not x[0].strip().startswith('http') and not x[0].strip().startswith('...')]
            #print(simr)
        else:
            simr = [x for x in simr if not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
    if len(simr) == 0:
        print ('No %s in the %d closet neighbors for word \"%s\"'%(flag, topk, word))
        print('Change the topk or change to another word')
        return
    simr = simr[:topk]
    if w2v == True:
        words = [x[0].strip() for x in simr]
    else:
        words = [x[1].strip() for x in simr]
    if word not in words:
        if w2v == True:
            simr.append((word, 1,0))
            words.append(word)
        else:
            simr.insert(0, (1.0, word))
            words.insert(0,word)
    if w2v == True:
        sims = {x[0]:x[1] for x in simr}
        #sims = {x[1]:x[0] for x in simr}
        
    for tupl in simr:
        if w2v == True:
            values.append(embedding.wv[tupl[0]])
        else:
            values.append(embedding.represent(tupl[1]))
    if wordset != None:
        for w in wordset:
            if w not in words:
                words.append(w)
                if w2v == True:
                    sims[w] = embedding.wv.similarity(w, word)
                    values.append(embedding.wv[w])
                else:
                    sims[w] = embedding.similarity(w, word)
                    values.append(embedding.represent(w))
    # print(words)
    cmap = plt.get_cmap('gnuplot')
    fitted = fit_tsne(values)
    if w2v == True:
        colors = [cmap(i) for i in np.linspace(0, 1, len(words))]
    else:
        colors = [cmap(i) for i in np.linspace(0, 1, topk+len(words))]
    annotations = plot_word(word, words, fitted, colors, sims)
    plt.show()
    
    
def plot_words(word1, words, fitted, cmap, sims, d, size=50):
    plt.axis('off')
    annotations = []
    for i in range(len(words)):
        pt = fitted[i]
        ww= words[i]
        sizing = sims[words[i]] * size
        plt.text(pt[0], pt[1], ww, color=d, size=int(sizing))
    return annotations

def plot_two_words(w1, w2, embedding, topk=10, flag = '', w2v=False):
    topn = 300
    values1 = []
    values2 = []
    if w2v == True:
        simr1 = embedding.most_similar(positive=[w1], topn = topk)
        simr2 = embedding.most_similar(positive=[w2], topn = topk)
        if (w1, 1.0) not in simr1:
            simr1.append((w1, 1.0))
        if (w2, 1.0) not in simr2:
            simr2.append((w2, 1.0))
        for (w, sim) in simr1:
            values1.append(embedding.wv[w])
        for (w, sim) in simr2:
            values2.append(embedding.wv[w])
    else:
        simr1 = embedding.closest(w1, topn)
        simr2 = embedding.closest(w2, topn)
        if flag == 'adj':
            simr1 = [x for x in simr1 if nltk.pos_tag([x[1].strip()])[0][1] in adjs and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
            simr2 = [x for x in simr2 if nltk.pos_tag([x[1].strip()])[0][1] in adjs and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
        elif flag == 'nn':
            simr1 = [x for x in simr1 if nltk.pos_tag([x[1].strip()])[0][1] in nns and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
            simr2 = [x for x in simr2 if nltk.pos_tag([x[1].strip()])[0][1] in nns and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
        else:
            if w2v == True:
                simr = [x for x in simr if not x[0].strip().startswith('http') and not x[1].strip().startswith('...')]
            else:
                simr1 = [x for x in simr1 if not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
                simr2 = [x for x in simr2 if not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
        if w1 not in [x[1].strip() for x in simr1]:
            simr1.insert(0, (1.0, w1))
        if w2 not in [x[1].strip() for x in simr1]:
            simr2.insert(0, (1.0, w2))
        simr1 = simr1[:topk]
        simr2 = simr2[:topk]    
        if len(simr1) == 0:
            print ('No %s in the %d closet neighbors for word \"%s\"'%(flag, topk, w1))
            print('Change the topk or change to another word')
            simr1 = [(0, '')]
        if len(simr2) == 0:
            print ('No %s in the %d closet neighbors for word \"%s\"'%(flag, topk, w2))
            print('Change the topk or change to another word')
            simr2 = [(0, '')]
        for (sim, w) in simr1:
            values1.append(embedding.represent(w))
        for (sim, w) in simr2:
            values2.append(embedding.represent(w))
    values = values1 + values2
    fitted = fit_tsne(values)
    fitted1 = fitted[:len(simr1),:]
    fitted2 = fitted[len(simr1):,:]
    max1 = np.max(fitted1, axis = 0)
    max2 = np.max(fitted2, axis = 0)
    min1 = np.min(fitted1, axis = 0)
    min2 = np.min(fitted2, axis = 0)
    dx = max1[0] - min2[0]
    dy = max1[1] - min2[1]
    plt.scatter(fitted[:,0], fitted[:, 1],  alpha=0)
    if w2v == True:
        words1 = [x[0].strip() for x in simr1]
        sims1 = {x[0]:x[1] for x in simr1}
    else:
        words1 = [x[1].strip() for x in simr1]
        sims1 = {x[1]:x[0] for x in simr1}
    # if w1 not in words1:
    #     if w2v == True:
    #         sims1[w1] = 1.0
    #         words1.append(w1)
    cmap = plt.get_cmap('gnuplot')
    colors = [cmap(i) for i in np.linspace(0, 1, topk)]
    annotations1 = plot_words(w1, words1, fitted1, colors, sims1, 'g')
    
    if w2v == True:
        words2 = [x[0].strip() for x in simr2]
        sims2 = {x[0]:x[1] for x in simr2}
    else:
        words2 = [x[1].strip() for x in simr2]
        sims2 = {x[1]:x[0] for x in simr2}
    # if w2 not in words2:
    #     if w2v == True:
    #         sims2[w2] = 1.0
    #         words2.append(w2)
    annotations2 = plot_words(w2, words2, fitted2+ [2*dx, 2*dy],  colors, sims2, 'b')
    plt.show()
    plt.close()
    

    

def plot_days(w1, w2, embeddings, topk=10, flag = '', w2v=False):
    topn = 300
    values = []
    plt.close('all')
    def _filter(embedding):
        values1 = []
        values2 = []
        if w2v == True:
            simr1 = embedding.most_similar(positive=[w1], topn = topn)
            simr2 = embedding.most_similar(positive=[w2], topn = topn)
            if (w1, 1.0) not in simr1:
                simr1.append((w1, 1.0))
            if (w2, 1.0) not in simr2:
                simr2.append((w2, 1.0))
            for (w, sim) in simr1:
                values1.append(embedding.wv[w])
            for (w, sim) in simr2:
                values2.append(embedding.wv[w])
        else:
            simr1 = embedding.closest(w1, topn)
            simr2 = embedding.closest(w2, topn)
            if flag == 'adj':
                simr1 = [x for x in simr1 if nltk.pos_tag([x[1].strip()])[0][1] in adjs and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
                simr2 = [x for x in simr2 if nltk.pos_tag([x[1].strip()])[0][1] in adjs and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
            elif flag == 'nn':
                simr1 = [x for x in simr1 if nltk.pos_tag([x[1].strip()])[0][1] in nns and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
                simr2 = [x for x in simr2 if nltk.pos_tag([x[1].strip()])[0][1] in nns and not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
            else:
                if w2v == True:
                    simr = [x for x in simr if not x[0].strip().startswith('http') and not x[1].strip().startswith('...')]
                else:
                    simr1 = [x for x in simr1 if not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
                    simr2 = [x for x in simr2 if not x[1].strip().startswith('http') and not x[1].strip().startswith('...')]
            if w1 not in [x[1].strip() for x in simr1]:
                simr1.insert(0, (1.0, w1))
            if w2 not in [x[1].strip() for x in simr1]:
                simr2.insert(0, (1.0, w2))
            simr1 = simr1[:topk]
            simr2 = simr2[:topk]    
            if len(simr1) == 0:
                print ('No %s in the %d closet neighbors for word \"%s\"'%(flag, topk, w1))
                print('Change the topk or change to another word')
                simr1 = [(0, '')]
            if len(simr2) == 0:
                print ('No %s in the %d closet neighbors for word \"%s\"'%(flag, topk, w2))
                print('Change the topk or change to another word')
                simr2 = [(0, '')]
            for (sim, w) in simr1:
                values1.append(embedding.represent(w))
            for (sim, w) in simr2:
                values2.append(embedding.represent(w))
        return simr1, simr2, values1, values2
    
    simrs = []
    count = 0
    for we in embeddings:
        simr1, simr2, v1, v2 = _filter(we)
        values = v1 + v2
        count += 1
        x = "1"+str(len(embeddings)) + str(count)
        plt.subplot(x)
        fitted = fit_tsne(values)
        fitted1 = fitted[:len(simr1),:]
        fitted2 = fitted[len(simr1):,:]
        max1 = np.max(fitted1, axis = 0)
        max2 = np.max(fitted2, axis = 0)
        min1 = np.min(fitted1, axis = 0)
        min2 = np.min(fitted2, axis = 0)
        dx = max1[0] - min2[0]
        dy = max1[1] - min2[1]
        plt.scatter(fitted[:,0], fitted[:, 1],  alpha=0)
        if w2v == True:
            words1 = [x[0].strip() for x in simr1]
            sims1 = {x[0]:x[1] for x in simr1}
        else:
            words1 = [x[1].strip() for x in simr1]
            sims1 = {x[1]:x[0] for x in simr1}

        cmap = plt.get_cmap('gnuplot')
        colors = [cmap(i) for i in np.linspace(0, 1, topk)]
        annotations1 = plot_words(w1, words1, fitted1+[count * dx, 0], colors, sims1, 'g', 30)

        if w2v == True:
            words2 = [x[0].strip() for x in simr2]
            sims2 = {x[0]:x[1] for x in simr2}
        else:
            words2 = [x[1].strip() for x in simr2]
            sims2 = {x[1]:x[0] for x in simr2}
        # if w2 not in words2:
        #     if w2v == True:
        #         sims2[w2] = 1.0
        #         words2.append(w2)
        annotations2 = plot_words(w2, words2, fitted2+ [count * dx, 2*dy],  colors, sims2, 'b', 30)
    plt.show()
    plt.close() 
        

def plot_sims_dis(wordset, embeddings):
    diff = []
    plt.close('all')
    for we in embeddings:
        diff.append(we.represent('she') - we.represent('he'))
    gender_dir = np.average(np.array(diff), 0)
    allsims = []
    for w in wordset:
        sims = []
        for we in embeddings:
            sims.append(we.represent(w).dot(gender_dir))
        plt.plot(range(3), sims,label=w)
        allsims.append(sims)
    plt.ylabel('he                she',fontsize=20)
    plt.xlabel('time line')
    plt.legend(loc='upper right')
    plt.show()   
    plt.close()
    
    ##### Other version####
def plot_fig(word, topk, f50, m50):
    plt.close('all')
    f, ax = plt.subplots(1, 2, subplot_kw=dict(projection='polar'), figsize=(10,10))
    x = f50.closest(word, topk)
    ax[0].axis('off')
    ax[0].scatter(0,0, s=20**2, color='red')
    ax[0].annotate(word, (0,0.05))
    ax[0].set_title('top%d neighbors for "%s" in %s'%(topk, word, 'feminist'), color = 'b')
    for (ind,(sim, w)) in enumerate(x):
            ax[0].scatter(np.pi*ind/len(x)*2, 1-sim, color='b')
            if w != word:
                ax[0].annotate(w, (np.pi*ind/len(x)*2, 1-sim+0.02), color = 'b')
    x = m50.closest(word, topk)
    ax[1].axis('off')
    ax[1].scatter(0,0, s=20**2, color='red')
    ax[1].annotate(word, (0,0.05))
    for (ind,(sim, w)) in enumerate(x):
            ax[1].scatter(np.pi*ind/len(x)*2, 1-sim, color='g')
            ax[1].annotate(w, (np.pi*ind/len(x)*2-0.04, 1-sim-0.03),color='g')
    ax[1].set_title('top%d neighbors for "%s" in %s'%(topk, word, 'misogynist'), color='g')
    f.subplots_adjust(wspace=0.3)
    plt.show()
    plt.close()
    
    
def plot_2w(w1, w2, f50, m50, topk=10):
    plt.close('all')
    sim1 = f50.closest(w1, topk)
    sim2 = f50.closest(w2, topk)
    fig = plt.figure(figsize=[8,8])
    ax1=fig.add_subplot(1,1,1)
    ax1.axis('off')
    def _plot(simres, word, cx, cy, co):
        ax1.scatter(cx, cy, s= 20**2, color='r')
        ax1.annotate(word, (cx, cy), color = co)
        for (indx, (sim, w)) in enumerate(simres):
            xd = (1-sim) * 1* math.cos(float(indx)/len(simres)*np.pi*2)
            yd = (1-sim) * 1* math.sin(float(indx)/len(simres)*np.pi*2)
            ax1.scatter(cx-xd, cy+yd, color=co)
            if w != word:
                ax1.annotate(w, (cx-xd+0.01, cy+yd-0.02), color = co)
    _plot(sim1, w1, 1, 0, 'g')
    _plot(sim2, w2, -1, 0, 'b')
    plt.show()
    plt.close()
    
def plot_2w_c(w1, w2, f50, topk=10):
    plt.close('all')
    f, ax = plt.subplots(1, 1, subplot_kw=dict(projection='polar'), figsize=(10,10))
    x = f50.closest(w1, topk)
    y = f50.closest(w2, topk)
    ax.axis('off')
    ax.scatter(1,0, s=20**2, color='red')
    ax.scatter(-1,0, s=20**2, color='red')
    ax.annotate(w1, (1,0.02), color='blue' , fontsize=20)
    ax.annotate(w2, (-1,0.02), color = 'green', fontsize = 20)
    for (ind,(sim, w)) in enumerate(x):
            ax.scatter(np.pi*ind/len(x)*2 + 1, 1-sim, color='b')
            if w != w1:
                ax.annotate(w, (np.pi*ind/len(x)*2 + 1, 1-sim+0.02), color='blue')
    for (ind,(sim, w)) in enumerate(y):
            ax.scatter(np.pi*ind/len(y)*2 - 1, 1-sim, color='g')
            if w != w2:
                ax.annotate(w, (np.pi*ind/len(y)*2 - 1, 1-sim+0.02), color = 'g')
    plt.show()
    plt.close()
```
